```python
# 5ms 1700K

import numpy as np
import matplotlib.pyplot as plt

# 상수 정의
C = 12.011
H = 1.00784

total_exitMassFlowRate = 3.502178e-3 # g/s

molar_masses = {
    'Styrene': C * 8 + H * 8,
    'Ethylbenzene': C * 8 + H * 10,
    'Toluene': C * 7 + H * 8,
    'C2H2': C * 2 + H * 2,
    'C2H4': C * 2 + H * 4,
    'CH4':  C + H * 4,
    'C6H6': C * 6 + H * 6,
    'C6H5C2H': C * 8 + H * 6,
    'C10H8': C * 10 + H * 8,
    'C12H8': C * 12 + H * 8,
    'C16H10': C * 16 + H * 10,
    'C14H10': C * 14 + H * 10,
    'BIN1B': C * 20 + H * 10,
    'BIN1A': C * 20 + H * 16
}

initialStyrene = 0.4614907
finalStyrene = 0.000002290319

mStyrene = initialStyrene - finalStyrene

convertStyreneCarbon = 8 * total_exitMassFlowRate * (mStyrene / molar_masses['Styrene'])

initialEthylbenzene = 0.202496
finalEthylbenzene = 0.0002122352

mEthylbenzene = initialEthylbenzene - finalEthylbenzene

convertEthylbenzeneCarbon = 8 * total_exitMassFlowRate * (mEthylbenzene / molar_masses['Ethylbenzene'])

initialToluene = 0.2439952
finalToluene = 0.002278515

mToluene = initialToluene - finalToluene

convertTolueneCarbon = 7 * total_exitMassFlowRate * (mToluene / molar_masses['Toluene'])

totalConvertCarbon = convertStyreneCarbon + convertEthylbenzeneCarbon + convertTolueneCarbon

mCH4 = 0.08517366
mC2H2 = 0.06781145
mC2H4 = 0.005814934
mC6H6 = 0.1937291
mC10H8 = 0.03715822
mC6H5C2H = 0.01593384
mC12H8 = 0.007079991
mBIPHENYL = 0.009258989
mC14H10 = 0.02082504
mC16H10 = 0.00651131
mC18H10 = 0.003175179
mBIN1B = 0.3288935
mBIN1A = 0.09719923

CH4Carbon = total_exitMassFlowRate * (mCH4 / molar_masses['CH4'])
C2H2Carbon = 2 * total_exitMassFlowRate * (mC2H2 / molar_masses['C2H2'])
C2H4Carbon = 2 * total_exitMassFlowRate * (mC2H4 / molar_masses['C2H4'])
C6H6Carbon = 6 * total_exitMassFlowRate * (mC6H6 / molar_masses['C6H6'])
C6H5C2HCarbon = 8 * total_exitMassFlowRate * (mC6H5C2H / molar_masses['C6H5C2H'])
C10H8Carbon = 10 * total_exitMassFlowRate * (mC10H8 / molar_masses['C10H8'])
C12H8Carbon = 12 * total_exitMassFlowRate * (mC12H8 / molar_masses['C12H8'])
C16H10Carbon = 16 * total_exitMassFlowRate * (mC16H10 / molar_masses['C16H10'])
C14H10Carbon = 14 * total_exitMassFlowRate * (mC14H10 / molar_masses['C14H10'])
BIN1BCarbon = 20 * total_exitMassFlowRate * (mBIN1B / molar_masses['BIN1B'])
BIN1ACarbon = 20 * total_exitMassFlowRate * (mBIN1A / molar_masses['BIN1A'])

selCH4 = (CH4Carbon / totalConvertCarbon) * 100
selC2H2 = (C2H2Carbon / totalConvertCarbon) * 100
selC2H4 = (C2H4Carbon / totalConvertCarbon) * 100
selC6H6 = (C6H6Carbon / totalConvertCarbon) * 100
selC6H5C2H = (C6H5C2HCarbon /totalConvertCarbon) * 100
selC10H8 = (C10H8Carbon / totalConvertCarbon) * 100
selC12H8 = (C12H8Carbon / totalConvertCarbon) * 100
selC14H10 = (C14H10Carbon / totalConvertCarbon) * 100
selC16H10 = (C16H10Carbon / totalConvertCarbon) * 100
selBIN1B = (BIN1BCarbon / totalConvertCarbon) * 100
selBIN1A = (BIN1ACarbon / totalConvertCarbon) * 100

totalSel = selBIN1A + selBIN1B + selC2H2 + selC2H4 + selC6H6 + selCH4 + selC14H10 + selC10H8 + selC6H5C2H + selC12H8 + selC16H10

print(selCH4)
print(selC2H2)
print(selC2H4)
print(selC6H6)
print(selC6H5C2H)
print(selC10H8)
print(selC12H8) #나프탈렌 
print(selC14H10)
print(selC16H10)
print(selBIN1B)
print(selBIN1A)
print(totalSel)
```
