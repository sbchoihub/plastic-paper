```python
import numpy as np
import matplotlib.pyplot as plt

# 상수 정의
C = 12.011
H = 1.00784

total_exitMassFlowRate = 3.502178e-3 #g/s

molar_masses = {
    'Styrene': C * 8 + H * 8,
    'Ethylbenzene': C * 8 + H * 10,
    'Toluene': C * 7 + H * 8,
    'C2H2': C * 2 + H * 2,
    'C2H4': C * 2 + H * 4,
    'C2H6': C * 2 + H * 6,
    'C3H6': C * 3 + H * 6,
    'CH4':  C + H * 4,
    'C4H8': C * 4 + H * 8,
    'C5H10': C * 5 + H * 10,
    'C6H6': C * 6 + H * 6,
    'C6H14': C * 6 + H * 14
}

def calculate_conversion(initial, final):
    return (initial - final) / initial * 100

def calculate_selectivities(flow_dict, initial_C6H14, final_C6H14):
    mStyrene = initial_Styrene - final_Styrene
    mEthylbenzene = initial_Ethylbenzene - final_Ethylbenzene
    mToluene = initial_Toluene - final_Toluene
    
    Styrene_carbon = 8 * total_exitMassFlowRate * (mStyrene / molar_masses['Styrene'])
    Ethylbenzene_carbon = 8 * total_exitMassFlowRate * (mEthylbenzene / molar_masses['Ethylbenzene'])
    Toluene_carbon = 7 * total_exitMassFlowRate * (mToluene / molar_masses['Toluene'])
    total_carbon = Styrene_carbon + Ethylbenzene_carbon + Toluene_carbon

    carbon_numbers = {
        'CH4': 1 * total_exitMassFlowRate * (flow_dict['CH4'] / molar_masses['CH4']),
        'C3H6': 3 * total_exitMassFlowRate * (flow_dict['C3H6'] / molar_masses['C3H6']),
        'C2H2': 2 * total_exitMassFlowRate * (flow_dict['C2H2'] / molar_masses['C2H2']),
        'C2H4': 2 * total_exitMassFlowRate * (flow_dict['C2H4'] / molar_masses['C2H4']),
        'C2H6': 2 * total_exitMassFlowRate * (flow_dict['C2H6'] / molar_masses['C2H6']),
        'C4H8_25': 4 * total_exitMassFlowRate * (flow_dict['C4H8_25'] / molar_masses['C4H8']),
        'C4H8_95': 4 * total_exitMassFlowRate * (flow_dict['C4H8_95'] / molar_masses['C4H8']),
        'C5H10_60': 5 * total_exitMassFlowRate * (flow_dict['C5H10_60'] / molar_masses['C5H10']),
        'C5H10_92': 5 * total_exitMassFlowRate * (flow_dict['C5H10_92'] / molar_masses['C5H10'])
    }

    selectivities = {key: val / total_carbon * 100 for key, val in carbon_numbers.items()}
    return selectivities, total_carbon

# 각 조건 입력값 정의
conditions = {
    '10cm': {
        'initial': 0.1790354,
        'final': 0.0001963348,
        'flow': {
            'C2H2': 0.009106949, 'C2H4': 0.1039785, 'C2H6': 0.00571584,
            'C3H6': 0.01999033, 'CH4': 0.03025541, 'C4H8_25': 2.52e-4,
            'C5H10_60': 5.36e-6, 'C5H10_92': 9.57e-6, 'C4H8_95': 1.98e-4
        }
    },
    '20cm': {
        'initial': 0.1790354,
        'final': 0.000007884703,
        'flow': {
            'C2H2': 0.011077, 'C2H4': 0.1051944, 'C2H6': 0.004896843,
            'C3H6': 0.01552831, 'CH4': 0.0326882, 'C4H8_25': 6.8e-5,
            'C5H10_60': 2.43e-6, 'C5H10_92': 4.31e-6, 'C4H8_95': 5.34e-5
        }
    },
    '30cm': {
        'initial': 0.1790354,
        'final': 0.0000003140263,
        'flow': {
            'C2H2': 0.01277729, 'C2H4': 0.1055805, 'C2H6': 0.004290245,
            'C3H6': 0.01225389, 'CH4': 0.03443814, 'C4H8_25': 3.88e-5,
            'C5H10_60': 1.78e-6, 'C5H10_92': 3.17e-6, 'C4H8_95': 3.05e-5
            
        }
    },
    '40cm': {
        'initial': 0.1790354,
        'final': 0.00000002827655,
        'flow': {
            'C2H2': 0.01392926, 'C2H4': 0.1055794, 'C2H6': 0.003947699,
            'C3H6': 0.01035962, 'CH4': 0.03548467, 'C4H8_25': 0.00002899144,
            'C5H10_60': 1.49e-6, 'C5H10_92': 2.65e-6, 'C4H8_95': 2.29e-5
        }
    }   
    
}

# 계산 수행
results = {}
for name, values in conditions.items():
    conversion = calculate_conversion(values['initial'], values['final'])
    selectivities, _ = calculate_selectivities(values['flow'], values['initial'], values['final'])
    total_selectivity = sum(selectivities.values())
    results[name] = {
        'conversion': conversion,
        'selectivities': selectivities,
        'total_selectivity': total_selectivity
    }

# 시각화용 데이터 구성
x = [10, 20, 30, 40]
data1 = np.array([results['10cm']['selectivities']['C2H4'],
                  results['20cm']['selectivities']['C2H4'],
                  results['30cm']['selectivities']['C2H4'],
                  results['40cm']['selectivities']['C2H4']])
data5 = np.array([results['10cm']['selectivities']['C3H6'],
                  results['20cm']['selectivities']['C3H6'],
                  results['30cm']['selectivities']['C3H6'],
                  results['40cm']['selectivities']['C3H6']])
data3 = np.array([results['10cm']['selectivities']['C2H2'],
                  results['20cm']['selectivities']['C2H2'],
                  results['30cm']['selectivities']['C2H2'],
                  results['40cm']['selectivities']['C2H2']])
data2 = np.array([results['10cm']['selectivities']['CH4'],
                  results['20cm']['selectivities']['CH4'],
                  results['30cm']['selectivities']['CH4'],
                  results['40cm']['selectivities']['CH4']])
data4 = np.array([results['10cm']['selectivities']['C2H6'],
                  results['20cm']['selectivities']['C2H6'],
                  results['30cm']['selectivities']['C2H6'],
                  results['40cm']['selectivities']['C2H6']])
data6 = np.array([results['10cm']['selectivities']['C4H8_25'],
                  results['20cm']['selectivities']['C4H8_25'],
                  results['30cm']['selectivities']['C4H8_25'],
                  results['40cm']['selectivities']['C4H8_25']])
data7 = np.array([results['10cm']['selectivities']['C4H8_95'],
                  results['20cm']['selectivities']['C4H8_95'],
                  results['30cm']['selectivities']['C4H8_95'],
                  results['40cm']['selectivities']['C4H8_95']])
data8 = np.array([results['10cm']['selectivities']['C5H10_60'],
                  results['20cm']['selectivities']['C5H10_60'],
                  results['30cm']['selectivities']['C5H10_60'],
                  results['40cm']['selectivities']['C5H10_60']])
data9 = np.array([results['10cm']['selectivities']['C5H10_92'],
                  results['20cm']['selectivities']['C5H10_92'],
                  results['30cm']['selectivities']['C5H10_92'],
                  results['40cm']['selectivities']['C5H10_92']])
data10 = np.array([results['10cm']['conversion'],
                  results['20cm']['conversion'],
                  results['30cm']['conversion'],
                  results['40cm']['conversion']])

# 시각화
plt.figure(figsize=(10, 6))

plt.bar(x, data1, color='b', width = 3, label = '$\\rm{C}_2\\rm{H}_4$')
plt.bar(x, data2, color='black', bottom = data1, width = 3, label = '$\\rm{CH}_4$')
plt.bar(x, data3, color='g', bottom = data1 + data2, width = 3, label = '$\\rm{C}_2\\rm{H}_2$')
plt.bar(x, data4, color='orange', bottom = data1 + data2 + data3, width = 3, label = '$\\rm{C}_2\\rm{H}_6$')
plt.bar(x, data5, color='c', bottom = data1 + data2 + data3 + data4, width = 3, label = '$\\rm{C}_3\\rm{H}_6$')
plt.bar(x, data6, color='gray', bottom = data1 + data2 + data3 + data4 + data5, width = 3, label = '$\\rm{C}_4\\rm{H}_{8}$')
plt.bar(x, data7, color='gray', bottom = data1 + data2 + data3 + data4 + data5 + data6, width = 3)#, label = '$\\rm{C}_5\\rm{H}_{10}$')
plt.bar(x, data8, color='gold', bottom = data1 + data2 + data3 + data4 + data5 + data6 + data7, width = 3, label = '$\\rm{C}_5\\rm{H}_{10}$')
plt.bar(x, data9, color='gold', bottom = data1 + data2 + data3 + data4 + data5 + data6 + data7 + data8, width = 3)#, label = '$\\rm{C}_5\\rm{H}_{10}$')

plt.xlim(0, 50)
ticklabel = ['0.1', '0.2', '0.3', '0.4']
plt.xticks(x, ticklabel, fontsize=17)
plt.yticks(fontsize=17)
plt.xlabel('z [m]', fontsize=20)
plt.ylabel('Selectivity [%]', fontsize=20)
plt.ylim(0, 105)
plt.grid(True, which='both', linestyle='--', axis='y')
ax = plt.gca()

ax.set_axisbelow(True)

# Conversion Line
ax2 = plt.gca().twinx()
line, = ax2.plot(x, data10, color='r', marker='s', linestyle='-', label='Conversion')
ax2.set_ylabel('Conversion [%]', fontsize=20)
plt.yticks(fontsize=17)
ax2.set_ylim(0, 105)



# 기존 핸들/레이블
handles1, labels1 = ax.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()  # [line], ['Conversion']

# 전체 핸들 및 라벨 결합
all_handles = handles1 + [line]
all_labels = labels1 + labels2

# 레이블 이름을 기준으로 원하는 순서 지정
desired_order = ['Conversion', 
                 '$\\rm{CH}_4$',
                 '$\\rm{C}_2\\rm{H}_2$', 
                 '$\\rm{C}_2\\rm{H}_4$',  
                 '$\\rm{C}_2\\rm{H}_6$', 
                 '$\\rm{C}_3\\rm{H}_6$',
                 '$\\rm{C}_4\\rm{H}_{8}$',
                 '$\\rm{C}_5\\rm{H}_{10}$']

# 순서대로 정렬
ordered_handles = [all_handles[all_labels.index(label)] for label in desired_order]
ordered_labels = desired_order

# 범례 그리기
ax.legend(ordered_handles, ordered_labels, loc='upper right', bbox_to_anchor=(1.4, 1.0), prop={'size': 14})

plt.savefig('./fig/1d_hexane_10ml_Conversion & Selectivity.pdf', bbox_inches='tight')
plt.show()

C2H4_20 = total_exitMassFlowRate * 0.1051944
print(C2H4_20)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

flow_rates = ['10cm', '20cm', '30cm', '40cm']
species = ['CH4', 'C3H6', 'C2H2', 'C2H4', 'C2H6', 'C4H8_25', 'C5H10_60', 'C5H10_92', 'C4H8_95']

# 전치 형태로 만들 딕셔너리
data = {}

# 각 flow rate에 대한 데이터 채우기
for rate in flow_rates:
    col_data = {}
    col_data['Conversion'] = results[rate]['conversion']
    col_data['total_selectivity'] = results[rate]['total_selectivity']
    for sp in species:
        col_data[sp] = results[rate]['selectivities'].get(sp, 0.0)
    data[rate] = col_data

# DataFrame 생성 (Component가 index)
df = pd.DataFrame(data)
df = df.round(2)  # 모든 값 소수점 둘째 자리까지

# PDF 저장
pdf_filename = './fig/Conversion & Selectivity flow rate table.pdf'
with PdfPages(pdf_filename) as pdf:
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.axis('tight')
    ax.axis('off')
    
    table = ax.table(cellText=df.values,
                     rowLabels=df.index,
                     colLabels=df.columns,
                     cellLoc='center',
                     loc='center')

    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1, 1.2)  # 셀 간격 살짝 키움
    table.auto_set_column_width(list(range(df.shape[1] + 1)))  # 모든 열 자동 너비 조정

    pdf.savefig(fig, bbox_inches='tight')
    plt.close(fig)

# 표 출력
print(df)
```
